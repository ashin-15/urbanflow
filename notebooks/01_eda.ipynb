# Exploratory Data Analysis (EDA)

This notebook conducts a thorough exploratory analysis of the Smart City Traffic Dataset. We analyze distributions, temporal traffic patterns, weather effects, and feature correlations to guide our modeling pipeline.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set aesthetics
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'font.size': 12,
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'figure.titlesize': 18
})

data_path = "../data/raw/smart_city_traffic_data.csv"
output_dir = "../reports/figures"

df = pd.read_csv(data_path)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

## 1. Traffic Demand Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['traffic_demand'], kde=True, bins=50, color='#1f77b4')
plt.title("Traffic Demand Distribution (Vehicles/Hour)", pad=15)
plt.xlabel("Traffic Demand")
plt.ylabel("Frequency")
plt.show()
```

## 2. Hourly Traffic Analysis

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='hour', y='traffic_demand', errorbar=('ci', 95), color='#2ca02c', linewidth=2)
plt.title("Hourly Traffic Demand Profile (with 95% Confidence Interval)", pad=15)
plt.xlabel("Hour of Day")
plt.ylabel("Mean Traffic Demand")
plt.xticks(range(0, 24))
plt.show()
```

## 3. Weather Impact Analysis

```python
plt.figure(figsize=(10, 6))
weather_order = df.groupby('weather_condition')['traffic_demand'].mean().sort_values(ascending=False).index
sns.barplot(data=df, x='weather_condition', y='traffic_demand', order=weather_order, hue='weather_condition', legend=False, palette='viridis')
plt.title("Mean Traffic Demand by Weather Condition", pad=15)
plt.xlabel("Weather Condition")
plt.ylabel("Mean Traffic Demand")
plt.show()
```

## 4. Correlation Analysis

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(12, 10))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True, linewidths=0.5)
plt.title("Numerical Feature Correlation Heatmap", pad=20)
plt.show()
```

## 5. Advanced Feature Engineering Analysis

To achieve a higher $R^2$, we engineered several advanced features such as cyclical temporal encodings, weather intensity compounds, and spatio-temporal target encodings. Here we visualize the correlation of these new engineered features with `traffic_demand`.

```python
# Apply new feature engineering to raw data to see correlations
import sys
sys.path.append('..')
from src.features import FeatureEngineer

fe = FeatureEngineer()
fe.fit(df)
df_trans = fe.transform(df)

# Select some new advanced features
advanced_features = [
    'traffic_demand', 'temp_wind_interaction', 'rainfall_humidity_interaction', 
    'vehicle_lane_ratio', 'rush_weekend_interaction', 'hour_location_encoded'
]

plt.figure(figsize=(10, 8))
corr_adv = df_trans[advanced_features].corr()
sns.heatmap(corr_adv, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True, linewidths=0.5)
plt.title("Advanced Feature Correlation Heatmap", pad=20)
plt.show()
```
